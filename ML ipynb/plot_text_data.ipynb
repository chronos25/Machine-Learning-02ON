```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
def plot_data(x,y):
    plt.plot(x,y,linestyle='',marker='x',color='r',label='Training data')
    plt.xlabel('Population in city in 10,000s')
    plt.ylabel('Population in $10,000s')
```

```python
data =np.loadtxt(open("ex1data1.txt","r"),delimiter=",")
x=data[:,0] # taking matrix of column 0
y=data[:,1] # taking matrix of column 1
m=len(y) #number of training examples & number of data in x & y & data are same

plt.figure()
plot_data(x,y) 
plt.show()
```
