```python
m=df2['vote_count'].quantile(0.50) # 50% the minimum votes required to be listed in the chart.

'''
#if(df2['vote_count']>=m):
    df2['vote_count']
    q=df2.copy().loc[df2['vote_count']>=m]
    q.shape
'''   
qualified_movies= df2.copy().loc[df2['vote_count']>=m]
#qualified_movies.shape
```

```python
def weighted_rating(x,m=m,c=c):
    v=x['vote_count']
    R=x['vote_average']
    imdb_score = (v/(v+m)*R)+(m/(m+v)*c)
    return (imdb_score)
```

```python
qualified_movies['score']=qualified_movies.apply(weighted_rating,axis=1)
#qualified_movies.head()
```

```python
qualified_movies=qualified_movies.sort_values('score',ascending=False)
qualified_movies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
df2['overview']=df2['overview'].fillna('')
tfidf_matrix = tfidf.fit_transform(df2['overview'])
#tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim=linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
indices=pd.Series(df2.index, index=df2['original_title']).drop_duplicates()
#indices
```

```python
def recommend(title,cosine_sim=cosine_sim):
    ind=indices[title]
    sim_score=list(enumerate(cosine_sim[ind]))
    sim_score=sorted(sim_score,key=lambda x: x[1],reverse=True)
    sim_score=sim_score[1:10]
    movie_indices=[i[0] for i in sim_score]
    return df2['original_title'].iloc[movie_indices]
```

```python
recommend('The Dark Knight')
```

```python
recommend('Clean')
```
